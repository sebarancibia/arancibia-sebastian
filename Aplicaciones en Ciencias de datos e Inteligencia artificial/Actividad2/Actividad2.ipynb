```python
def optimizar_datos(x, y):
    
    x = x.reshape(-1, 1)  
    y = y.reshape(-1, 1)

    X = np.hstack([np.ones((x.shape[0], 1)), x])  
    
    beta = np.linalg.pinv(X.T @ X) @ X.T @ y
    
    return beta[0], beta[1] 

intercepto, pendiente = optimizar_datos(X, y)
print("Intersección:", intercepto)
print("Pendiente:", pendiente)
```

```python
def grafica_regLineal(x, y, intercepto, pendiente):
    
    plt.figure(figsize=(20, 12))
    plt.scatter(x, y, color='royalblue', alpha=0.6, label="Datos") 
    
    x_linea = np.linspace(x.min(), x.max(), 1000)  
    y_linea = intercepto + pendiente * x_linea  
    plt.plot(x_linea, y_linea, color='red', linewidth=2, label="Recta de regresión") 
    plt.xlim(-0.1,3.1)
    plt.ylim(6.5,15)
    
    
    plt.title("Ajuste de Regresión Lineal", fontsize=16)
    plt.xlabel("Variable independiente (x)", fontsize=14)
    plt.ylabel("Variable dependiente (y)", fontsize=14)

    
    plt.tick_params(axis='both', which='major', labelsize=12)
    plt.legend(fontsize=12)
    plt.show()


grafica_regLineal(X, y, intercepto, pendiente)
```
