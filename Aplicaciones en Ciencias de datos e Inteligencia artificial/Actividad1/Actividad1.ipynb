```python
import pandas as pd
```

```python
#archivo = open("datos/ejemplo_data.csv","r", encoding= "UTF-8")
df = pd.read_csv("../datos/ejemplo_data.csv", sep=",", header=0)
#df = pd.read_csv("../datos/ejemplo_data.csv", sep=",", header=0, index_col=0)

df
```

```python
print(df.dtypes)
print()
print(df.info())
```

```python
df["ID"] = df["ID"].astype(int)
df["Activo"] = df["Activo"].astype(bool)


df.info()
```

```python
df["Nombre"] = df["Nombre"].astype("category")

df.info()
```

```python
def convertir(val):
    nueva_data = val.replace("$", "").replace(",", "").replace("%", "")

    return float(nueva_data)
```

```python
df["2016"] = df["2016"].apply(convertir)
df["2017"] = df["2017"].apply(convertir)
df["Crecimiento"] = df["Crecimiento"].apply(convertir)

df["2016"] = df["2016"].astype(float)
df["2017"] = df["2017"].astype(float)
df["Crecimiento"] = df["Crecimiento"].astype(float)

print(df.info())
df
```

```python
df["Unidades"] = pd.to_numeric(df["Unidades"],errors="coerce")
df["Unidades"] = df["Unidades"].astype(int, errors="ignore")

print(df)

df.info()
```

```python
# 2.2

df2 = pd.read_csv("../datos/ecommerce_data.csv", sep=",", header=0, encoding= "UTF-8", on_bad_lines="skip")

df2
```

```python
print(df2.dtypes)

print()

print(df2.info())
```

```python
def convertir2(val):
    nueva_data = val.replace("C", "").replace("A", "")

    return nueva_data
```

```python
# No podia transformar la columna InvoiceNo a int porque tiene C en algunas posiciones, por lo que aplicare replace
df2["InvoiceNo"] = df2["InvoiceNo"].apply(convertir2)

df2["InvoiceNo"] = df2["InvoiceNo"].astype(int)
df2["Description"] = df2["Description"].astype(str)

df2.info()
```

```python
df2["Quantity"] = df2["Quantity"].astype(int)
df2["UnitPrice"] = df2["UnitPrice"].astype(float)

df2.info()
```

```python
# Separar en dos columnas y unir a DF
fecha = df2["InvoiceDate"].str.split(" ",expand=True)
fecha.columns = ["Fecha", "Hora"]

df2 = pd.concat([df2, fecha], axis=1)
df2
```

```python
# Eliminar columna
f2 = df2.drop(columns = ["InvoiceDate"]) 
df2
```

```python
precio = df2["Quantity"] * df2["UnitPrice"]
precio = pd.DataFrame(precio)
precio
precio.columns = ["Precio Total"]

df2 = pd.concat([df2, precio], axis=1)
```

```python
df2.to_csv("Actividad 1 , Punto 2.2.csv")
```
