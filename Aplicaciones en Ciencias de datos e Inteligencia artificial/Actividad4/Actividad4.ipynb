2.1 Implementacion y visualizacion de k-means en 2D

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("kmeans.1.csv")
df
```

```python
plt.figure()
plt.scatter(df["A"], df["B"])
#plt.plot(df["A"], df["B"],"k.")
plt.grid(True)
```

```python
from sklearn.cluster import KMeans
```

```python

kmeans = KMeans(n_clusters=2)
#kmeans = KMeans(n_clusters=4)
kmeans.fit(df)
```

```python
centroide = kmeans.cluster_centers_
print(centroide)
print()
puntos = kmeans.labels_
print(puntos)

print()
distancia = kmeans.inertia_
print(distancia)
```

```python
plt.scatter(df['A'], df['B'], cmap='viridis', marker='o')
plt.scatter(centroide[:, 0], centroide[:, 1], color='red', marker='X')
```
